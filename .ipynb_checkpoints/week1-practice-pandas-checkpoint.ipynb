## 1. 연도(Year)별 총 생산량

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'D:/2019_02_학부_강의자료/인공지능/02주차-실습/과제01/'
```

```python
df = pd.read_csv(path + 'energy.csv')
```

```python
df['total'] = df['Solar Energy'] + df['Bio (Gas)'] + df['Wind Power'] + df['Hydro Power'] + df['Small Hydro'] \
+ df['Fuel Cell'] + df['Ocean Energy']
```

```python
grouped_by_year = df.groupby(by = ['Year'])
```

```python
result = grouped_by_year[['total']].sum()
result
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'

plt.grid()
plt.title('연도별 총 생산량')
plt.xlabel('연도')
plt.ylabel('총 생산량')
#plt.plot(result)
plt.bar(result.index, result['total'], color = 'orange')
plt.show()
```

## 2. 도시(Region)별 총 생산량

```python
grouped_by_region = df.groupby(by = ['Region'])
```

```python
result = grouped_by_region[['total']].sum()
result
```

```python
plt.figure(figsize = (16, 8))
plt.grid()
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.title('도시별 총 생산량')
plt.xlabel('지역')
plt.ylabel('총 생산량')
plt.bar(result.index, result['total'], color='springgreen')
plt.show()
```

# 3. 신재생 에너지 종류별 총 생산량

```python
result = df.sum()[df.columns[4:-2]]
print(result)
```

```python
plt.grid()
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.title('신재생 에너지 종류별 총 생산량')
plt.xlabel('에너지')
plt.ylabel('총 생산량')
plt.bar(result.index, result.values, color='yellow')
plt.show()
```
